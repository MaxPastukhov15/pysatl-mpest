```python
from em_algo.types import *

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

import numpy as np
```

```python
from examples.utils import *

results = open_results("quick_test")
```

```python
for i, result in enumerate(results):

    table = []

    for single_solver_result in result.results:
        table += [
            [
                d.model.name,
                tuple(d.model.params_convert_from_model(d.params)),
                d.prior_probability,
                single_solver_result.solver.optimizer.name
            ]
            for d in single_solver_result.result.result
        ]
    df = pd.DataFrame(table, columns=["name", "params", "prior_probability", "optimizer"])
    pd.set_option("display.expand_frame_repr", False)
    #print(df)
    print(f"test â„–       : {i}")
    print(f"samples count: {len(result.test.problem.samples)}")
    for single_solver_result in result.results:
        print("__________________________________________________")
        print()
        print(df[df["optimizer"] == single_solver_result.solver.optimizer.name])
        print(f"time         : {single_solver_result.time:.3f}")
        print(f"steps        : {single_solver_result.steps}")
        print(f"error        : {single_solver_result.result.error}")
    fig, ax1 = plt.subplots()
    ax1.set_xlabel("x")

    sns.histplot(result.test.problem.samples, color="lightgrey", label="samples")
    ax2 = ax1.twinx()
    ax2.set_ylabel("p(x)")

    min_data = min(result.test.all_data)
    max_data = max(result.test.all_data)

    X = list(np.linspace(min_data, max_data, int(np.abs(max_data - min_data) * 512)))
    ax2.plot(X, [result.test.true_mixture.pdf(x) for x in X], "--", color="green", label="target")
    for single_solver_result in result.results:
        ax2.plot(
            X,
            [single_solver_result.result.result.pdf(x) for x in X],
            label=single_solver_result.solver.optimizer.name,
        )
    ax2.legend()
    plt.show()
    print()
    print()
```
